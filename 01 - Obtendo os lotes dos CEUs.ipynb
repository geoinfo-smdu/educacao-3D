# Obtendo os lotes fiscais dos CEUs

Esse é o primeiro objetivo para podermos quantificar os elementos físicos construídos e vegetados: precisamos conhecer os limites geográficos da área de estudo. Para isso vamos utilizar a camada de CEUs disponibilizada pelo erviço WFS do GeoSampa

## Obtendo os pontos dos CEUs por serviço WFS

```python
from owslib.wfs import WebFeatureService
import json
import geopandas as gpd
from shapely.geometry import shape
```

```python
url_wfs = 'http://wfs.geosampa.prefeitura.sp.gov.br/geoserver/geoportal/wfs'

wfs = WebFeatureService(url_wfs, version='2.0.0')
```

```python
# [operation.name for operation in wfs.operations]
```

```python
# Listando todas as camadas disponíveis
list(wfs.contents)
```

```python
response = wfs.getfeature(typename='geoportal:equipamento_educacao_ceu', outputFormat='application/json')
```

```python
ceus = json.loads(response.read())
```

```python
# ceus
```

```python
gdf = gpd.GeoDataFrame.from_features(ceus['features'])
gdf
```

```python
gdf.plot()
```

```python
from owslib.wms import WebMapService

url_wms = 'http://wms.geosampa.prefeitura.sp.gov.br/geoserver/geoportal/wms'

wms = WebMapService(url_wms, version='1.3.0')
```

```python
lotes = []
gdf_lotes = gdf

for k, r in gdf.iterrows():
    c = r.geometry.coords[0]
    response = wms.getfeatureinfo(layers = ['lote_cidadao'],
                          format = 'image/png',
                          info_format = 'application/json',
                          srs = 'EPSG:31983',
                          bbox = (c[0] - 0.1, c[1] - 0.1, c[0] + 0.1, c[1] + 0.1),
                          size = (1, 1), 
                          xy = (1, 1))
    lote = json.load(response)
    print(f"{len(lote['features'])} feições para o ID {k}")
    if len(lote['features']) > 0:
        lotes.append(shape(lote['features'][0]['geometry']))
#     break
```

```python
gpd.GeoDataFrame(geometry=lotes, crs=31983).to_file('resultados/lotes-wms.geojson', driver='GeoJSON')
```
